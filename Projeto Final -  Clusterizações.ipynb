```python
dfc = df.drop(['Cidade', 'Estado', 'Religião - 2010'], axis=1)
```

```python
sse={}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000)
    kmeans.fit(X=dfc)
    dfc["Clusters"] = kmeans.labels_
    sse[k] = kmeans.inertia_
```

```python
plt.figure(figsize=(10, 5))
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("dfc")
plt.show()
```
